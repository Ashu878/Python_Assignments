### 1. You are given two strings s and t. String t is generated by randomly shuffling string s and then adding one more letter at a random position. Return the letter that was added to t.

```python
import logging
```

```python
logging.basicConfig(filename = 'test.log', level = logging.DEBUG, format = '%(message)s %(error)s')
```

```python

class fd:
    """Class to find difference"""
    def __init__(self,first,second):
        try:
            self.first = first
            self.second = second
            if (type(self.first) != str) and (type(self.second) != str):
                raise Exception("Error raised")
        except Exception as e:
            print(e)
            logging.error(e)
            
    def find_the_difference(self):
        """Class object to find difference between words"""
        try:
            a = set(self.first)
            b = set(self.second)
            to_return = list(a.symmetric_difference(b))
            return " ".join(to_return)
        except Exception as e:
            print(e)
            
            
```

```python
a = fd('abcd','abcde')
```

### 2. Given a function that accepts unlimited arguments, check and count how many data types are in those arguments. Finally return the total in a list.

```python
def count_datatypes(*args):
    """Function to output data types"""
    try:
        main = [i for i in args]
        integers,string,boolean,lists,tuples,dics = 0,0,0,0,0,0
        for i in main:
            if type(i) == int:
                integers += 1
            if type(i) == str:
                string += 1
            if type(i) == bool:
                
                boolean += 1
            if type(i) == list:
                lists += 1
            if type(i) == tuple:
                tuples += 1
            if type(i) == dict:
                dics += 1
        return [integers,string,boolean,lists,tuples,dics]
    except Exception as  e:
        print(e)
        logging.error("Error outptu")
    
    
    
```

```python
count_datatypes(1, 45, "Hi",
               True) 
```

```python
count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23) 
```

```python
count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6]) 
```

### 3. A Fibonacci string is a precedence of the Fibonacci series. It works with any two characters of the English alphabet (as opposed to the numbers 0 and 1 in the Fibonacci series) as the initial items and concatenates them together as it progresses in a similar fashion as the Fibonacci series.



```python
def fibo(till,words):
    """Prog for fibo"""
    try:
        main = ""
        if type(words)  != list:
            raise Exception("Invalid input")
        for i in range(len(words)-1):
            a,b = words[i],words[i+1]
        
        for i in range(till):
            to_add = "{} ".format(a)
            a,b = b,a+b
            main += to_add
        return main
    except Exception as e:
        print(e)
        logging.error(e)
    
   
    
```

```python
fibo(3,['j','h'])
```

```python
fibo(5,['e','a'])
```

```python
fibo(6,['n','k'])
```

### 4. Given an integer between 0 and 26, make a variable (self.answer). That variable would be assigned to a string in this format:"nines:your answer, threes:your answer, ones:your answer" You need to find out how many ones, threes, and nines it would at least take for the number of each to add up to the given integer when multiplied by one, three or nine (depends).


```python
def one_three_nines(number):
    """Prog"""
    try:
        a = number//9
        nines = number%9
        
        b = nines//3
        threes = nines%3
        
        ones = threes//1
        return "Nines:{}, Threes:{}, Ones:{}".format(a,b,ones)
    except Exception as e:
        print(e)
        logging.error(e)
        
        
```

```python
one_three_nines(15)
```

```python
one_three_nines(10)
```

```python
one_three_nines(22)
```
